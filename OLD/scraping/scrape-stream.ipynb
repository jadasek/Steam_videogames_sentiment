```python
from steam_review_scraper import get_review_count
from steam_review_scraper import search_game_id
from steam_review_scraper import get_game_review 
import pandas as pd

games = ['PUBG', 'Grand Theft Auto V', 'Borderlands 2', 'Red Dead Redemption 2', 'Detroit Become Human', 'Skyrim', 'Dark Souls 3', 'Portal 2', 'Terraria', 'The witcher 3', 'God of War']
game_ids = []
output = pd.DataFrame(columns=['user','playtime','user_url','post_date','helpfulness','review','recommend','early_access_review','game_id'])

for i in games:
    ids = search_game_id(i)
    game = ids.iloc[0]['id']
    game_ids.append(game)

print(game_ids)
for i in game_ids:
    reviews = get_game_review(i)
    reviews['game_id'] = i
    output = pd.concat([output,reviews])
```
